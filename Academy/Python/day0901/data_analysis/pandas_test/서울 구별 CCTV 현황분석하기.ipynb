```python
import pandas as pd ,os
os.listdir()
```

```python
CCTV_s=pd.read_excel("서울시 자치구 (연도별) CCTV 설치현황_241231.xlsx")
CCTV_s.head()
```

```python
CCTV_s=CCTV_s.drop(["순번"],axis=1) #axis= 1 열방향
CCTV_s.head()
```

```python
CCTV_s.isna()
```

```python
CCTV_s.fillna(value=0,inplace=True)
CCTV_s.isna()
```

```python
CCTV_s.info()
```

```python
pop_s=pd.read_excel('등록인구_20250826085221.xlsx',header=2,usecols='B,D')
# 2번부터(3번행) 출력, 사용하는 열은 B,D
pop_s.head()
```

```python
# 컬럼명 변경
pop_s.rename(columns={pop_s.columns[0]:'구별',pop_s.columns[1]:'인구수'},inplace=True)
pop_s.head()
```

```python
# CCTV가 많이 설치된 순서(총 계)대로 정렬
CCTV_s.sort_values(['총 계'], ascending=False)
```

```python
# CCTV 컬럼명 확인
CCTV_s.columns
```

```python
# 최근 증가율 추가
CCTV_s['최근 증가율']=(CCTV_s['2016년']+CCTV_s['2017년']+CCTV_s['2018년']+CCTV_s['2019년']+
                  CCTV_s['2020년']+CCTV_s['2021년']+CCTV_s['2022년']+CCTV_s['2023년']+CCTV_s['2024년']
                  )/(CCTV_s['2015년 이전 설치된 CCTV '])*100
CCTV_s.head()
```

```python
CCTV_s['최근 증가율'] = (CCTV_s['2017년']+CCTV_s['2018년']+CCTV_s['2019년']+CCTV_s['2020년']+CCTV_s['2021년']+CCTV_s['2022년']+CCTV_s['2023년'] +CCTV_s['2024년'] 
 ) / (CCTV_s['2015년 이전 설치된 CCTV '] + CCTV_s['2016년']) * 100
# 2015년이에 값이 없는 경우 때문에 2016년 값을 추가
CCTV_s
```

```python
# '최근 증가율' 기준으로 정렬
CCTV_s.sort_values('최근 증가율',ascending=False).head()
```

```python
# CCTV에서 '계' 행 삭제

CCTV_s.drop([0],inplace=True)
CCTV_s
```

```python
#pop_s의 '소계' 행 삭제

pop_s.drop([0],inplace=True)
pop_s.head()
```

```python
# CCTV의 컬럼, '구분'-> '구별' 이름 변경

CCTV_s.rename(columns={CCTV_s.columns[0]: '구별'},inplace=True)
CCTV_s.columns
```

```python
# 컬럼 변경 확인 
CCTV_s.head()
```

```python
# CCTV의 구별 컬럼의 값을 확인, 및 중복 제거(unique)
CCTV_s['구별'].unique()
```

```python
# pop의 구별 컬럼의 값을 확인, 및 중복 제거(unique)
pop_s['구별'].unique()
```

```python
# '중 구' -> '중구'
CCTV_s.loc[2,'구별']='중구'
CCTV_s.loc[2,'구별']
```

```python
# CCTV, pop merge
data_result=pd.merge(CCTV_s,pop_s,on='구별') # 구별을 기준으로 merge
data_result
```

```python
# 불필요한 줄 삭제
del data_result['2015년 이전 설치된 CCTV ']
del data_result['2016년']
del data_result['2017년']
del data_result['2018년']
del data_result['2019년']
del data_result['2020년']
del data_result['2021년']
del data_result['2022년']
del data_result['2023년']
del data_result['2024년']

data_result
```

```python
# index 를 '구별'로 수정 

data_result.set_index('구별',inplace=True)
data_result.head()
```

```python
import numpy as np
np.corrcoef(data_result['인구수'],data_result['총 계']) # 상관계수
```
