```python
import pandas as pd
from scipy import stats
from statsmodels.formula.api import ols, glm
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm


# scipy
```

```python
# 파일 읽기

# 파일의 구분자가 ';'로 되어있음. -> ';' 를 기준으로 읽기
# header 가 0인 경우 굳이 작성할 필요는 없으나, 타인이 읽기에 가독성이 좋아 작성
red_df = pd.read_csv('data/winequality-red.csv', sep = ';', header = 0, engine = 'python')
white_df = pd.read_csv('data/winequality-white.csv', sep = ';', header = 0, engine= 'python')
```

```python
# 파일 확인
red_df.head()
```

```python
# 파일 확인
white_df.head()
```

```python
# white와 합치기전에 구분을 위해 type 컬럼생성
red_df.insert(0,column='type',value='red')
```

```python
# 컬럼 추가 확인
red_df.head()
```

```python
# 합치기전에 구분을 위해 type 컬럼생성
white_df.insert(0,column='type',value='white')
```

```python
# 컬럼 추가 확인
white_df.head()
```

```python
# df 가 몇행,열인지 확인
white_df.shape
```

```python
# df 가 몇행,열인지 확인
red_df.shape
```

```python
# concat으로 두 df를 합침
# 합친 df가 몇행,열인지 확인
wine= pd.concat([red_df,white_df])
wine.shape
```

```python
# wine의 정보확인
wine.info()
```

```python
# wine의 컬럼명에 있는 공백을 '_'로 변경
wine.columns=wine.columns.str.replace(' ','_')
wine.columns
```

```python
# wine의 통계 요약
wine.describe()
```

```python
# wine에 있는 'quality' 컬럼의 값이 어떤게 있는지 종류 확인
sorted(wine.quality.unique())
```

```python
# wine의 'quality' 컬럼의 종류와 종류별 갯수 확인
wine.quality.value_counts()
```

```python
# red와 white 그룹간 quality 통계
wine.groupby('type')['quality'].describe()
```

```python
# red와 white 그룹간 quality 통계, 원하는 요소만 출력
wine.groupby('type')['quality'].agg(['mean','std','max'])
```

```python
#레드 와인과 화이트와인 등급을 비교한 t검정의 결과(유의미X)
red_wine_quality = wine.loc[wine['type'] == 'red', 'quality']
white_wine_quality = wine.loc[wine['type']== 'white','quality']
# p-value 가 0.05 이하로 통계적으로 유의

stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
```

```python
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
# 회귀분석 : 두 변수사이의 인과관계 파악
regression_result = ols(Rformula, data=wine).fit()

regression_result.summary()
```

```python
sample1=wine[wine.columns.difference(['quality','type'])]
sample1=sample1[0:5][:]
sample1
```

```python
sample1_predict=regression_result.predict(sample1)
sample1_predict
```

```python
sns.set_style('dark')
sns.histplot(red_wine_quality, kde = True, color='red', label='red wine') # kde 커널 밀도
sns.histplot(white_wine_quality, kde = True, label='white wine')
plt.title("Quality of Wine Type")
plt.legend()
plt.grid()
plt.show()
```
