```python
# 1. CSV 파일 읽기
f = open("data_analysis/subway2507.csv", "r", encoding="utf-8")
rideSum = {}
alightSum = {}

for line in f.readlines()[1:]:  # 헤더 건너뛰기
    line = line.replace("\n", "").split(",")
    
    station = line[2]            # 역명
    ride_cnt = int(line[3].strip('"'))      # 승차총승객수
    alight_cnt = int(line[4].strip('"'))    # 하차총승객수

    if station in rideSum:
        rideSum[station] += ride_cnt
        alightSum[station] += alight_cnt
    else:
        rideSum[station] = ride_cnt
        alightSum[station] = alight_cnt

f.close()

# 2. 리스트로 변환
name = []
ride = []
alight = []

for k in rideSum:
    name.append(k)
    ride.append(rideSum[k])
    alight.append(alightSum[k])

# 3. NumPy 배열로 변환
import numpy as np

name = np.array(name)
ride = np.array(ride)
alight = np.array(alight)

# 4. 승차보다 하차가 많은 역 출력
print(name[ride < alight])
```
