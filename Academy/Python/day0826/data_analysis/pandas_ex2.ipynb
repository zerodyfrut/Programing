```python
df=df.replace({',':''},regex=True) # 정규식 패턴 ','를 ''로 변경
# regex : 정규표현식 여부
df=df.apply(pd.to_numeric,errors='ignore') 
# apply : 모든 것들에 대해 적용
# error='ignore' : 에러는 무시한다. 적용 안되는 것은 그냥 무시
df.head()
# 거래금액의 ',' 삭제 확인
```

```python
#특정 조건의 열을 출력
df[df['전용면적(㎡)']>130].head()
# head()를 통해 5개만 출력
```

```python
# 특정 조건의 열에서 특정 컬럼만 출력
df['거래금액(만원)'][df['전용면적(㎡)']>130].head()
# 전용면적이 130보다 큰 행중에서 거래금액 열만 출력
# 너무 많으니 head() 적용
```

```python
# 조건이 여러 개일 경우 
# &를 사용
df[(df['전용면적(㎡)'].round(2)>130) & (df['거래금액(만원)']<15000)].head()
```

```python
# or조건
# |를 사용
df[(df['전용면적(㎡)']>130) | (df['거래금액(만원)']<15000)].head()
```

```python
# 행과 열의 특정부분 출력(slice)
# loc [slice, [컬럼명]]
# 얘는 slice인데 해당갯수+1개를 출력하네..
df.loc[:3,['시군구','단지명','전용면적(㎡)','거래금액(만원)']]
```

```python
# 행과 열의 특정부분 조건 출력

df.loc[:,['시군구','단지명','전용면적(㎡)','거래금액(만원)']][(df['거래금액(만원)']>5000) & (df['거래금액(만원)']<10000)]
```

```python
# 새로운 열 추가
# 없는 컬럼값을 사용하면 새로운 컬럼이 생성
df['면적대비금액']=df['거래금액(만원)']/df['전용면적(㎡)']
df.loc[:10,['거래금액(만원)','전용면적(㎡)','면적대비금액' ]]
```

```python
# 정렬
# sort_values(by '컬럼명') : 오름차순 정렬
# sort_values(by '컬럼명',ascending=False) : 내림차순 정렬

df.sort_values(by='거래금액(만원)',ascending=True).iloc[:10,[0,1,2,3,7]]
```
