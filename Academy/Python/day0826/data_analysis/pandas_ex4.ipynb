```python
date=pd.date_range('2025-01-01',periods=5,freq='YS')
# Pandas에서 날짜 범위를 생성하는 함수.
# pd.date_range('시작날짜',period=생성개수, freq='Y')
# freq=Y : 연단위(연말), YS, 연초, M : 월단위(월말), MS, 월초, Q: 분기말, D:일
print(date)

df=pd.DataFrame({
    'A':pd.Series(np.random.randint(0,10,size=5),index=date),
    'B':pd.Series(np.random.randint(0,10,size=5),index=date),
    'C':pd.Series(np.random.randint(0,10,size=5),index=date),
    'D':pd.Series(np.random.randint(0,10,size=5),index=date)
})

df
```

```python
# 열추가 : 기존에 없었던 열을 작성
df['E']=pd.Series(np.random.randint(0,10,size=5),index=date)
df
```

```python
# 행추가
# 기존에 없던 행을 loc를 통해 작성
# df.loc['2030-01-01']=[0,0,0,0,0]
df.loc['2030-01-01']=[0,0,0,0,0]
df
```

```python
# 새로운 열을 특정 위치에 추가하기
# insert(추가할 위치,컬럼명, 들어갈 값)
df.insert(2,'AA',pd.Series(np.random.randn(6),index=pd.date_range('1/1/2025',periods=6,freq='YS')))
df
```

```python
# 행 삭제
df.drop('2025-01-01',axis=0,inplace=True)
df
```

```python
#열 삭제
# df=df수정값 : inplace=True 와 동일한 효과
df=df.drop('E',axis=1)
df
```

```python
# replace(to_replace, value)
# 안의 값대신 저렇게 쓸수있는데, 별차이가 있나?

df=df.replace(to_replace=0.0,value=df.mean())
df
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])
```

```python
# DF확인
display(df1)
display(df2)
```

```python
# df 결합
# concat([df1,df2]) : 위아래로 결합
result=pd.concat([df1,df2])
result
```

```python
# DF 구분용 key 입력
# index의 값을 추가하는 것
result=pd.concat([df1,df2],keys=['x','y'])
result
```

```python
# key값은 멀티 인덱스로 적용됨
result.index
```

```python
# concat시 해당하는 index 값이 없는 셀은 nan 처리

df3 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'], 
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])

result=pd.concat([df1,df3],axis=1)
result
```

```python
# 공통된 index가 있는 값만 join 해서 출력
# inner join
result=pd.concat([df1,df3],axis=1, join='inner')
display(df1)
display(df3)
result
```

```python
# DF 생성 및 확인
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

display(left)
display(right)
```

```python
# 공통 컬럼 key 기준으로 합침
display(left)
display(right)
pd.merge(left,right,on='key')
```

```python
# 각 결합 출력
display(left)
display(right)

# left의 key값을 기준으로 합침
display(pd.merge(left,right,on='key',how='left'))
# right의 key값을 기준으로 합침
display(pd.merge(left,right,on='key',how='right'))
# 합칠때 index 값이 없는 셀은 nan 처리
display(pd.merge(left,right,on='key',how='outer'))
# 공통 컬럼 key 기준으로 합침
display(pd.merge(left,right,on='key',how='inner'))
```
