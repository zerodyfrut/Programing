```python
from selenium import webdriver
import time
import pandas as pd
from bs4 import BeautifulSoup

def CoffeeBean_store(result):
    CoffeeBean_URL = "https://www.coffeebeankorea.com/store/store.asp"
    wd = webdriver.Chrome()

    for i in range(1, 150):
        wd.get(CoffeeBean_URL)
        time.sleep(1) # 웹페이지 연결할 동안 1초 대기
        
        try:
            #js함수 실행
            wd.execute_script("storePop2(%d)" % i)
            time.sleep(1)
            
            html = wd.page_source # JS 함수가 수행된 페이지의 소스코드
            
            soupCB = BeautifulSoup(html, 'html.parser') # 소스를 읽고, Html document로 바꿔줌
            store_name_h2 = soupCB.select("div.store_txt > h2") # class가 store_txt 인 div 안의 h2태그를 가져옴
            store_name = store_name_h2[0].string # 하나만 있어도 list로 가져옴
            print(store_name)
            
            store_info = soupCB.select("div.store_txt > table > tbody > tr > td")
            store_address_list = list(store_info[2]) # td태그에 들어있는게 여럿이라 list로 반환
            store_address = store_address_list[0] # td태그 안에 들어있는 첫번째것
            store_phone = store_info[3].text #4번째 td 태그
            
            result.append([store_name, store_address,store_phone])
        except:
            continue
    return

def main():
    result = [] # 여러지점 정보를 담을 list
    print('CoffeeBean crawling >>>>>>>>>>>>>>>>>>>>>>>>>>>>>')
    
    CoffeeBean_store(result)        
    print(">>>")
    
    cb_df = pd.DataFrame(result, columns=('store', 'address', 'phone'))
    cb_df.to_csv("CoffeeBean_store.csv", encoding='utf-8', mode='w', index=True)
    
if __name__ == '__main__':
    main()
```
