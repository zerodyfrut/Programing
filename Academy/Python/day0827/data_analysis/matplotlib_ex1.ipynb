```python
# list 생성
x=[1,4,9,16,25,36,49,64]
y=[i for i in range(1,9)]

# 자동으로 list-> np.array로 바꿔서 활용

plt.plot(x,y) # 선 그래프
plt.xlabel('x') # x축 라벨
plt.ylabel('y') # y축 라벨
plt.title('matplotlib 연습') # 그래프 제목
plt.show() # 그래프 그리기
```

```python
t=np.arange (0,12,0.01)  # 0 ~ 12전 가지 간격은 0.01 => 대략 1200개 정도
print(t)
y=np.sin(t) # t에 저장된 값으로 sin 값
print(y)
```

```python
# 그래프 크기설정 및 sin/cos 그래프, 격자

plt.figure(figsize=(6,4)) # 그래프가 나올 사이즈
plt.plot(t,y,label='sin') # t범위간 y에 저장된 sin 값으로 그래프
plt.plot(t,np.cos(t),color='r',label='cos') #t범위간 cos(t) 값으로 붉은색 그래프
# plt.plot(t,np.tan(t),color='g',label='tan') 

plt.grid() # 격자
plt.title('Sin and Cos')
plt.legend() # 빈공백에 항목별 구분선 설명, plot에 설정한 label 값
plt.show()
```

```python
# 한번에 여러 개의 그래프 그리기
t=[i for i in range(7)]
y=[1,2,5,8,9,5,3,]

fig=plt.figure(figsize=(10,6)) 

ax1= fig.add_subplot(2,2,1) # 2행 2열의 첫번째
ax1.plot(t,y,color='b')
ax2= fig.add_subplot(2,2,2)
ax2.plot(t,y,color='k',linestyle='--')
ax3= fig.add_subplot(2,2,3)
ax3.plot(t,y,color='r',linestyle=':',marker='>')
ax4= fig.add_subplot(2,2,4)
ax4.plot(t,y,color='y',linestyle='-.',marker='o',markerfacecolor='m')

plt.show()
```

```python
# 그래프에 점찍기, 화살표 출력, 포인트에 txt 출력

x = [1,2,3]
y = [2,4,8]

plt.title("꺽은선 그래프")
plt.xlabel('X 축')
plt.ylabel('Y 축')

plt.plot(x,y,marker='o')  

for idx, txt in enumerate(y):
    #(x좌표,y좌표,보여줄문자,ha(수평정렬),color)
    plt.text(x[idx],y[idx]+0.2,txt, ha='center', color='blue')  
    # enumerate의 index를 통해 접근한 위치에 value값 출력, 스타일 지정

# 차트 작성
plt.plot(1.25, 3, '*') # 점
# annotate : xy(화살표 시작 위치), xytext(annotate 문자열 위치), shrink(화살촉과 차트사이 간격)
plt.annotate('annotate', xy=(1.25, 3), xytext=(1.5, 4.5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
# bar chart

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]

# 범례 레이블
bar_labels = ['red', 'blue', '_red', 'orange'] # _를 사용하면,동일 레이블이 반복 출력되지 않음
# bar_labels =['apple', 'blueberry', 'cherry', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange'] # tab:붙이면 조금 연해지는 듯

plt.bar(fruits, counts, label=bar_labels, color=bar_colors)

plt.ylabel('fruit supply')
plt.title('Fruit supply by kind and color')
plt.legend(title='Fruit color')

plt.show()
```

```python
# bar 생성 및 출력

x=np.arange(4)
years=['2002','2021','2022','2023']
values=[100,400,900,1000]

plt.bar(x,values, color=['r','g','b'], width=0.3)
plt.xticks(x,years)

plt.show()
```

```python
# 수평 bar 생성 및 출력

x=np.arange(4)
years=['2002','2021','2022','2023']
values=[100,400,900,1000]

plt.barh(x,values, color=['r','g','b'], height=0.3)
plt.yticks(x,years)

plt.show()
```

```python
# pie

ratio=[34,32,16,18]
labels=['Apple','banana','Melon','Grape']
plt.pie(ratio, labels=labels, autopct='%.1f%%')

plt.show()
```

```python
# 특정 pie에 간격을 두고 출력하기

ratio=[34,32,16,18]
labels=['Apple','banana','Melon','Grape']
explode=[0,0.1,0,0.1]
plt.pie(ratio, labels=labels, autopct='%.1f%%',startangle=260,counterclock=False,explode=explode)

plt.show()
```

```python
# pie 색상 꾸미기, pie간 동일한 간격두고 간격간 색상채우기

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.5, 'edgecolor': 'g', 'linewidth': 5}
            # 부채꼴 너비, 테두리색상, 테두리 너비
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=90, counterclock=True, colors=colors, wedgeprops=wedgeprops)
# counterclock=False 시계 방향 순서
# startangle는 부채꼴이 그려지는 시작 각도
plt.show()
```

```python
# Stacked bar chart : 막대 그래프를 겹쳐서 보여주는 형태
# 한 항목당 여러 값을 누적해서 하나의 막대로 표시

months=['Jan','Feb','Mar']
apple=np.array([10,15,20])
banana=np.array([5,8,12])
cherry=np.array([2,4,6])

plt.bar(months, apple, label='사과')
plt.bar(months, banana, bottom=apple, label='바나나') # apple 위에 쌓기
plt.bar(months, cherry, bottom=apple+banana, label='체리') # apple +banana 위에 쌓기

plt.ylabel('판매량')
plt.title('월별 과일 판매량 (Stacked Bar Chart)')
plt.legend()

plt.show()
```

```python
# 값의 크기에따라 색상이 정해지는 colorbar

import matplotlib.pyplot as plt
import numpy as np

# np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()
plt.show()
```

```python
# 박스 플롯 : 5가지 요약 통계 - 최소값, 1사분위, 중앙값, 3사분위, 최대값 , o -> 이상치(범위 밖의 값)
# 데이터 분포 파악, 이상치 탐지

import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
# np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000) # 평균, 표준편파, 표본수
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c])
ax.set_ylim(-10.0, 10.0) # y축 범위를 고정
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```
