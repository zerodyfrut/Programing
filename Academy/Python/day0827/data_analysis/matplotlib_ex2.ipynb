```python
import matplotlib.pyplot as plt, numpy as np, os,pandas as pd
import matplotlib.font_manager as fm
os.chdir(r'D:\새 폴더\Programing\Academy\Python\day0827')
```

```python
# read csv

f=pd.read_csv('data_analysis/data/students_data.csv')
df=pd.DataFrame(f)
df
```

```python
# 한글 글씨체 설정
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin': # 애플
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows': # window
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
# 국어성적 bar 차트 그리기

plt.bar(df['성명'],df['국어'],color='tab:green')
plt.title('학생별 국어성적')
plt.yticks([i for i in range(0,101,10)])

plt.show()
```

```python
# 국어성적 bar 차트 그리기 - 점수별 색상 추가

colors=[]
for score in df['국어']:
    if score>=90:
        colors.append('blue')
    elif score>=80:
        colors.append('green')
    elif score>=70:
        colors.append('yellow')
    else:
        colors.append('red')

plt.bar(df['성명'],df['국어'],color=colors)
plt.title('학생별 국어성적')
plt.yticks([i for i in range(0,101,10)])

plt.show()
```

```python
# 학생들의 수학, 과학 점수를 bar 차트로 출력, 이 때 y축을 공유하는 subplot으로 그림

fig=plt.figure(figsize=(12,4))
ax1=fig.add_subplot(1,2,1)
ax1.bar(df['성명'],df['수학'],color='tab:blue')
ax1.set_title('학생별 수학성적')

ax2=fig.add_subplot(1,2,2)
ax2.bar(df['성명'],df['과학'],color='tab:green')
ax2.set_title('학생별 과학성적')

ax1.set_yticks([i for i in range(0,101,10)])
ax2.set_yticks([i for i in range(0,101,10)])

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(12,4))
ax1.bar(df['성명'], df['수학'],color="g")
ax1.set_title('학생별 수학 점수')

ax2.bar(df['성명'], df['과학'])
ax2.set_title('학생별 과학 점수')
plt.show()
```

```python
# 두개를 한번에 띄우는 bar
import numpy as np

x = np.arange(len(df))  # 학생 위치
width = 0.3  # 막대 너비

plt.bar(x - width/2, df['수학'], width, label='수학', color='g')
plt.bar(x + width/2, df['과학'], width, label='과학', color='b')

plt.xticks(x, df['성명'])  # x축 라벨
plt.ylabel('점수')
plt.title('학생별 수학/과학 성적 비교')
plt.legend()
plt.show()
```

```python
fig, ([[ax1, ax2], [ax3,ax4]]) = plt.subplots(2, 2,figsize=(12,8))
ax1.bar(df['성명'], df['국어'],width=0.7)
ax1.grid()
ax1.legend(['국어점수'])
ax1.set_title('학생별 국어 점수')
ax1.set_xlabel('성명')
ax1.set_ylabel('점수')

x = np.arange(df['성명'])
ax2.plot(x, df['영어'])
ax2.plot(x, df['수학'])
ax2.set_yticks([i for i in range(78,95,2)])
ax1.set_xlabel('성명')
ax1.set_ylabel('점수')

ax2.legend(['영어점수','수학점수'])
ax2.set_title('학생별 영어/수학 점수')

plt.show()

```

```python
fig, axes = plt.subplots(2,2,figsize=(15,10)) # 2행 2열
fig.suptitle('여러 개의 그래프') # 4개를 관리하는 전체 타이틀

# 첫번째 그래프 작성
axes[0,0].bar(df['성명'],df['국어'],label='국어점수')
axes[0,0].set_title('학생별 국어 점수')
axes[0,0].legend()
axes[0,0].set(xlabel='성명',ylabel='점수')
axes[0,0].set_facecolor('lightyellow') # 전경색
axes[0,0].grid(linestyle='--', linewidth=0.5)


# 두번째 그래프 작성
axes[0,1].plot(df['성명'],df['영어'],label='영어점수')
axes[0,1].plot(df['성명'],df['수학'],label='수학점수')
axes[0,1].set_title('학생별 영어/수학 점수')
axes[0,1].legend()
axes[0,1].set(xlabel='성명',ylabel='점수')


# 세번째 그래프 작성
axes[1,0].barh(df['성명'],df['과학'])
axes[1,0].set_title('학생별 과학점수')
axes[1,0].set(xlabel='점수',ylabel='성명')


# 네번째 그래프 작성
axes[1,1].plot(df['성명'], df['체육'], color='green', alpha=0.5)
axes[1,1].set(xlabel='성명',ylabel='점수')
axes[1,1].set_title('학생별 체육점수')

plt.show()
```
