```python
# inclass
#'서울'로 시작하는 역이름(중복 제거)
station_series = df['역명'][df['역명'].str.startswith('서울')]
unique_stations1 = station_series.drop_duplicates() 
# drop_duplicates() : 중복값 제거
print(unique_stations1)

unique_stations2 = station_series.drop_duplicates().reset_index(drop=True) #인덱스도 리셋
print(unique_stations2)
```
